```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to alist of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## Line Chart

#### MATLAB style input

```python
plt.plot(x_axis, y_axis)
plt.show()
```

#### Object Oriented input

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# or
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

#### Matlab version

```python
# Create the plot and add alabel for the legend.
plt.plot(x_axis, y_axis, marker = '*', color = 'blue', linewidth = 2, label = 'Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title
plt.title('PyBer Fare by Month')

# Add the legend.
plt.legend()
```

#### Object Oriented version

```python
# add anotations with object oriented approach
fig, ax = plt.subplots()

ax.plot(x_axis, y_axis, color = 'green', marker = 'D', linewidth = 2, label = 'Boston')

ax.set_title('PyBer Fare by Month')

ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

ax.set_ylim(0, 45)

ax.grid()

ax.legend()
```

## Bubble Chart 
### Matlab method

```python
# add 's = ""' as parameter to plt.scatter() to change size of markers
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
# make y_axis values bigger by looping over y_axis and multiplying by a number (3 in this case)
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(3 * data)
```

```python
# set 's' paramter to y_axis_larger to make the bubbles bigger
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
# use list comprehension to replace the new list and for loop above
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

### Scatter Plot Object Oriented Method

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

### Bubble Chart Object Oriented Method

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

```python
# make markers sky blue, change markers to 5 * data point, make color 20% transparent
# add black edge color to the circles, make the linewidth of the circles 2 points, legend for Boston,
# add title, switch axis so Fare ($) is x_axis, add limit to x axis, invert y axis to Jan on top
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[ i *5 for i in y_axis], alpha=0.8, edgecolor='k', linewidth=2, label='Boston')

ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_xlim(0,50)
ax.invert_yaxis()
ax.legend()
```

## Pie Chart

### Matlab Method

```python
# set x_axis (month) as label, y_axis (fare) as value of wedge
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# add percentages to wedges and 'explode' the biggest wedge 
# set tuple for percentage for each wedge to explode as a variable
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# set list of colors to use for wedges as variable 
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# make chart bigger
plt.subplots(figsize = (8, 8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct = '%.1f%%')
plt.show()
```

### Object Oriented Pie Char

```python
explode_values = (0,0,.2,0,0,0,.2,0,0,0,0,0,)
colors = ['lime', 'seagreen','turquoise', 'cornflowerblue', 'darkorchid', 'lightcoral', 'firebrick', 'bisque','goldenrod', 'lawngreen','cyan', 'teal']

fig, ax = plt.subplots()
#fig = plt.figure(figsize=(8,8))


ax.pie(y_axis, explode=explode_values, labels=x_axis, radius= 2, colors=colors, shadow=True, startangle=90,counterclock=False, autopct='%.1f%%')
plt.show()
```
