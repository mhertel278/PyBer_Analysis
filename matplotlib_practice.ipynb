```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to alist of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## Line Chart

#### MATLAB style input

```python
plt.plot(x_axis, y_axis)
#plt.show()
```

#### Object Oriented input

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# or
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

#### Matlab version

```python
# Create the plot and add alabel for the legend.
plt.plot(x_axis, y_axis, marker = '*', color = 'blue', linewidth = 2, label = 'Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title
plt.title('PyBer Fare by Month')

# Add the legend.
plt.legend()
```

#### Object Oriented version

```python
# add anotations with object oriented approach
fig, ax = plt.subplots()

ax.plot(x_axis, y_axis, color = 'green', marker = 'D', linewidth = 2, label = 'Boston')

ax.set_title('PyBer Fare by Month')

ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

ax.set_ylim(0, 45)

ax.grid()

ax.legend()
```

## Bar Charts

#### matlab version input

```python
# create the plot
plt.bar(x_axis, y_axis)
```

```python
# annotate the bar chart
plt.bar(x_axis, y_axis, color = "green", label= "Boston")

plt.xlabel("Date")
plt.ylabel('Fare($)')

plt.title('PyBer Fare per Month')

plt.ylim(0,45)

plt.legend()
```

### Horizontal Bar Chart

```python
plt.barh(x_axis, y_axis)
```

```python
# switch axes of horizontal bar chart
plt.barh(y_axis, x_axis)
```

```python
# redo horizontal bar chart with months on y axis, but invert the order to go Jan to Dec
plt.barh(x_axis, y_axis)
# use gca--'GET CURRENT AXES' method to select the y axis, and use invert method to invert
plt.gca().invert_yaxis()
```

```python
# redo horizontal chart with dates on y, y axis inverted, annotated, with legend for Boston, and colored magenta
# create plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")

# invert y axis
plt.gca().invert_yaxis()

# label the axes
plt.xlabel('Fare($)')
plt.ylabel('Date')

# set limit on x axis
plt.xlim(0, 45)

# create legend
plt.legend()
# create title
plt.title('PyBer Fare by Month')

```

### Object Oriented Approach for Bar Charts

```python
# create plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#  color black, and create legend for New York
fig, ax =plt.subplots()
ax.bar(x_axis, y_axis, color = "k", label = "New York")

ax.legend()

ax.set_title("PyBer Fare by Month")

ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

ax.set_ylim(0, 45)
```

```python
# make horizontal bar chart
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# switch axes of hor. bar chart
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# hor. bar chart with legend for Chicago, color cyan, fares on X axis, invert y axis to have Jan at top, annotate
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = 'c', label = "Chicago")

ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')

ax.invert_yaxis()

ax.set_title('PyBer Fares by Month')

ax.legend()

```
